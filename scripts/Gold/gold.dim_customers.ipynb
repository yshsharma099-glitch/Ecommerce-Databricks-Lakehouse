```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY customer_order_key) AS customer_key,
    customer_order_key,
    customer_id,
    customer_city,
    customer_state,
    zip_prefix AS zip_code
FROM workspace.silver.customers
"""
```

```python
df = spark.sql(query)

(
    df.write
    .mode("overwrite")
    .format("delta")
    .option("overwriteSchema","true")
    .saveAsTable("workspace.gold.dim_customers")
)

print(" SUCCESS: dim_customers created with validated columns.")
```
