```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY seller_id) AS seller_key,
    seller_id,
    seller_city,
    seller_state
FROM workspace.silver.sellers
"""
df = spark.sql(query)
```

```python
(
    df.write
    .mode("overwrite")
    .format("delta")
    .option("overwriteSchema", "true")
    .saveAsTable("workspace.gold.dim_sellers")
)
```
